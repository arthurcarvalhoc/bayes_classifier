```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
features =[ 
'Profile_mean',
'Profile_stdev',
'Profile_skewness',
'Profile_kurtosis',
'DM_mean',
'DM_stdev',
'DM_skewness',
'DM_kurtosis',
]
```

```python
data = pd.read_csv('HTRU2/HTRU_2.csv',header=None,names=features+['label'])
```

```python
data.head()
```

```python
data.label.value_counts().plot(kind='bar')
plt.title('Quantidade de exemplos por label')
```

```python
fig, ax = plt.subplots(2,1,figsize=(14,12))
data[features].boxplot(ax=ax[0])
ax[0].set_title('Distribution of Feature Values')

scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(data[features]))
y = data.label

data_scaled = pd.DataFrame(X)
data_scaled['label'] = y
data_scaled.columns = features + ['label']
data_scaled[features].boxplot(ax=ax[1])
ax[1].set_title('Distribution of Feature Values - Normalized Using Standard Scaler')
```

```python
sns.pairplot(data=data_scaled,
             palette=['red','orange'],
             hue="label",
             vars=features)

plt.suptitle("Pair Plot of HTRU2 Dataset ",fontsize=16)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show() 
```
